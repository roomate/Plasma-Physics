## Some computations on Astrophysical Plasma Physics

### Import modules

```python
import numpy as np
import matplotlib.pyplot as plt
```

In astrophysical plasma, it is well-known that there is multiple source of waves in the electricomagnetism spectrum, depending on the physical conditions considered.

Recall that under a static magnetic field $\mathbf{B_0}$, charged particles will tend to adopt a rotating movement with an axis of rotation parallel to $\mathbf{B_0}$. It is quite trivial to show that frequency is related to its amplitude $B$ via the formula: $\omega_i=\frac{ZeB_0}{m_ic} \ll \omega_e=\frac{eB_0}{m_e c}$ since $m_i \approx 2000m_e$. Moreover, Alfven wave propagates in the same direction as $\mathbf{B_0}$.

It propagates at Alfven velocity: $v_{\text{A}}^g=\frac{B_0}{\sqrt{\mu_0 \rho}}$. This is a group velocity (i.e. the actual moving speed of a wave packet), the phase velocity being: $v_{\text{A}}^p=\displaystyle\frac{v_{\text{A}}^g}{\sqrt{1 + \frac{v_{\text{A}}^g}{c^2}}}$. Since $v_A^g$ does not depend explicitly on $\omega$, the wave frequency, it is dispersionless.

To show this, consider the linear compressible MagnetoHydroDynamic (MHD) model. Put yourself in the plasma frame, such that the overall plasma velocity is null. The plasma is supposed to be inviscid ($\nu=0$), an ideal conductor ($\eta=0$) and under a strong static magnetic field $B_0\hat z$. Also, assume the plasma undergo a reversible and adiabatic process. This assumption provides a state equation: Laplace's law. Laplace's law simply states that $\frac{p}{\rho^\gamma}=\text{Cte} \iff \frac{d}{dt}\frac{p}{(mn)^\gamma}=0$, where $\rho=mn= \rho_e + \rho_i$ and $\gamma=\frac{C_p}{C_v}$ is the adiabatic coefficient. Hence, the ideal MHD equations are:
$$
\begin{array} {cc}
\partial_t n + \nabla \cdot (nv) = 0 & \text{(Mass conservation equation)}\\
mn(\partial_t v + (v \cdot \nabla) v) = -\nabla p + J \times B & \text{(Momentum conservation equation)}\\
\partial_t B = \nabla \times (v \times B) & \text{(Maxwell-Faraday equation)}\\
\frac{d}{dt}(\frac{p}{(nm)^\gamma}) & \text{(Laplace's Law)}.
\end{array}
$$

This form of Maxwell-Faraday equation unfolds from generalized Ohm's law in ideal conducting fluid and original Faraday equation: $\nabla \times E = -\partial_t B$. This set of equation are quite a good approximation for astrophysical plasma, where large-scale phenomena on the magnitude of $100$ km/h have the most impact. For small scale physics though, it does not yield physically consistent results, especially around magnetic reconnection zones. The freezing theorem, a natural consequence of the ideal MHD equations from above, does not prevail in those conditions.

Now, without loss of generality, take a linear perturbation of the magnetic field of the form: $B=B_0 \hat z + B_1= B_0 \hat z + \mathbf{B_1}e^{j(\omega t - k_{\perp} z - k_{\parallel} x)}$ with $|\mathbf{B_1}| \ll B_0$ and assume that $|\mathbf{v_1}| \ll 1, p_1 \ll p_0$ and $n_1 \ll n_0$.

After injection and equating orders, it yields:
$$
\begin{array} {cc}
\partial_t n_1 + n_0\nabla \cdot v_1 = 0 & \text{(Mass conservation equation)}\\
mn_0\partial_t v_1 = -\nabla p_1 + \frac1{\mu_0} (\nabla \times B_1) \times B_0 & \text{(Momentum conservation equation)}\\
\partial_t B_1 = \nabla \times (v_1 \times B_0) & \text{(Maxwell-Faraday equation)}\\
\partial_t p_1 = \frac{\gamma p_0}{mn_0} \partial_t n_1 & \text{(Laplace's Law)} \\
E=-v \times B & \text{(Ohm's law)}
\end{array}
$$

For the second equation, one used Maxwell-Ampère's law, stating in the low-frequency limit (neglicting the displacement current) that $\nabla \times B=\mu_0 J$. Note $C_s^2=\frac{\gamma p}{mn_0}=\sqrt{\beta \gamma} V_A$ the sound celerity. We would like to find relations between $\omega$, $k_\perp$ and $k_\parallel$ to have non-trivial solutions $\mathbf{B_1}$. After eliminating $p_1$ and $B_1$ with $v_1$, you arrive at the following linear equation:
$$
\underbrace{\begin{bmatrix}
-\omega^2 + (kV_A)^2 + k_\perp^2 C_s^2 & 0 & k_\perp k_\parallel C_s^2 \\
0 & -\omega^2 + k_\parallel V_A^2 & 0 \\
k_\perp k_\parallel C_s^2 & 0 & -\omega^2 + k_\parallel C_s^2
\end{bmatrix}}_{A \in \mathbb{R}^{3\times 3}} v_1 = 0
$$
for which $\text{Ker}(A) \neq \emptyset \iff \boxed{\text{det}(A) = 0}$. Each relation between $\omega$ and $k=(k_\perp, 0, k_\parallel)=k_\perp \hat x + k_\parallel \hat z$ such that $\text{det}(A)=0$ is called a **dispersion relation** and yields its own type of electromagnetic wave.

Just by looking at it, one can enumerates three solutions:
- If the middle term of the second column is null, then the whole column is null and obviously $\text{det}(A)=0$. It implies $\omega^2=kV_A$. This is called Alfven mode.
- If the first column and third column are colinears, then once again $\text{det}(A)=0$. It would implies that $\frac{-\omega^2 + (kV_A)^2 + k_\perp^2 C_s^2}{k_\perp k_\parallel C_s^2} = \frac{k_\perp k_\parallel C_s^2}{-\omega^2 + k_\parallel C_s^2} \iff (-\omega^2 + k_\parallel C_s^2)(-\omega^2 + (kV_A)^2 + k_\perp^2 C_s^2)=(k_\perp k_\parallel C_s^2)^2$. It admits two solutions: $\omega^2=\frac{k^2}2\big(V_A^2 + C_s^2 + \sqrt{(V_A^2 + C_s^2)^2 - 4(V_A^2 C_s^2\frac{k_\parallel^2}{k^2})} \big)$ and $\omega^2=\frac{k^2}2\big(V_A^2 + C_s^2 + \sqrt{(V_A^2 + C_s^2)^2 + 4(V_A^2 C_s^2\frac{k_\parallel^2}{k^2})} \big)$. Those are called respectively slow and fast magnetosonic modes.

To sum up, solutions are:
- **Alfven mode**: Let $\omega^2= k_\parallel^2 V_A^2$. They are low-frequency i.e. lower than the ion gyrofrequency. This mode does not provoke density fluctuations at first order solution since $v_1 \in \text{Ker}(A)=\text{Vect}(\hat y)$ and continuity equation in frequency domain yields: $n_1=n_0\frac{k\cdot v_1}{\omega}=0$. 
- **fast compressible magnetosonic mode**: Let $\omega^2=\frac{k^2}2\big(V_A^2 + C_s^2 + \sqrt{(V_A^2 + C_s^2)^2 - 4(V_A^2 C_s^2\frac{k_\parallel^2}{k^2})} \big)$.
- **slow compressible magnetosonic mode**: Let $\omega^2=\frac{k^2}2\big(V_A^2 + C_s^2 - \sqrt{(V_A^2 + C_s^2)^2 - 4(V_A^2 C_s^2\frac{k_\parallel^2}{k^2})} \big)$.

Recall the definition of the phase speed: $v_\phi\equiv\frac{\omega}{\text{Re}(k)}$ and group speed $v_g\equiv\frac{\partial \omega}{\partial k}$.

## Objective

Consider a spacecraft with relative speed $V_{\text{SW}}$ to the Solar Wind.

In general, such a spacecraft probing an electromagnetic wave with angular frequency $\omega_{\text{plasma}}$ will have to take into account its own speed in the measurement, since it inevitably causes a Doppler shift $\omega_{\text{Doppler}}=k\cdot V_{\text{SW}}$ in the overall spectrum. At the end, the spectrum $\omega_{\text{sc}}$ measured by a probe is the sum of both: $\omega_{\text{sc}}=\omega_{\text{plasma}} + \omega_{\text{Doppler}} = \omega_{\text{plasma}} + k\cdot V_{\text{SW}}$. Note that $k$ is determined by $\omega_{\text{plasma}}$ via the dispersion relation, the latter being dependent on the wave considered. As the above section hints, we will consider the three wave discussed: Alfven, slow and fast magnetosonic modes.

The objective of the next cells is to compare the influence of solar wind $V_{SW}$ with electromagnetic waves (Alfven, Slow and Fast) propagating along $\textbf{k}$. Note $\theta_{k-B_0}$ the angle between the perturbation propagation and the static magnetic field $\mathbf{B_0}$. Note $\theta_{k-V_{SW}}$ the angle between $\textbf{k}$ and $\mathbf{V_{SW}}$. Therefore, $k_\parallel=k\cos(\theta_{k-B_0})$.

Taylor's hypothesis states that doppler shift completely dominates the spectrum observed by the spacecraft: $\omega_{\text{Doppler}}=kV_{\text{SW}}\cos(\theta_{kV_{\text{SW}}}) \gg \omega_{\text{plasma}}$.

Here, one would like to see under which configurations a doppler shift prevail over the other. A colormap will display the ratio $\frac{\omega_{\text{wave}}}{\omega_{\text{Doppler}}}$.

### Physical considerations

- $V_A=100 \, \text{km/h}$,

- $V_{SW}= 400-800 \, \text{km/h}$,

- $\theta_{k-SW}=40°$,

- $\theta_{k-B_0} = 0 - 180 °$

### Alfven mode

$\frac{\omega_{\text{A}}}{\omega_{\text{Doppler}}}=\frac{kV_{\text{A}}\cos(\theta_{k-B_0})}{kV_{\text{SW}}\cos(\theta_{k-SW})}=\frac{V_{\text{A}}\cos(\theta_{k-B_0})}{V_{\text{SW}}\cos(\theta_{k-SW})}$

```python
#Number of entries
N = 100

#Alfven speed
Va = 100

#Solar wind array of speed
V_SW = np.linspace(400,800,N)

#Angle between magnetic field and wave
theta_kb = np.linspace(0,np.pi,N)
theta_kV = 40*2*np.pi/360
Beta = 2
gamma = 5/3
om_a = Va*np.abs(np.cos(theta_kb))

X,Y = np.meshgrid(V_SW,theta_kb)

cs = np.sqrt(Beta*gamma)*Va

om_dopp = V_SW*np.cos(theta_kV)

R = np.empty((N,N))

for i in range(N):
        R[i] = om_a[i]/om_dopp

plt.contourf(X,Y,R,50)
plt.title('Alfven wave')
plt.xlabel('Solar wind velocity $V_{SW}$')
plt.ylabel('angle k-B (in rad)')
plt.colorbar()
plt.show()
```

### Fast Magnetosonic mode

$\frac{\omega_{\text{F-MS}}}{\omega_{\text{Doppler}}}=\frac{\frac{k}{\sqrt{2}}\sqrt{\big(V_A^2 + C_s^2 + \sqrt{(V_A^2 + C_s^2)^2 - 4(V_A^2 C_s^2\frac{k_\parallel^2}{k^2})} \big)}}{kV_{\text{SW}}\cos(\theta_{k-SW})}=\frac{\frac{1}{\sqrt{2}}\sqrt{\big(V_A^2 + C_s^2 + \sqrt{(V_A^2 + C_s^2)^2 - 4(V_A^2 C_s^2\frac{k_\parallel^2}{k^2})} \big)}}{V_{\text{SW}}\cos(\theta_{k-SW})}$.

```python
N = 100
va  = 100
V_SW = np.linspace(400,800,N)
theta_kb = np.linspace(0,np.pi,N)
theta_kV = 40*2*np.pi/360
Beta = 2
gamma = 5/3

X,Y = np.meshgrid(V_SW,theta_kb)

cs = np.sqrt(Beta*gamma)*va

om_f = np.sqrt((cs**2 + va**2)/2*(1 + np.sqrt(1 - 4*np.cos(theta_kb)**2*va**2*cs**2/(cs**2 + va**2)**2)))

om_dopp = V_SW*np.cos(theta_kV)

R_f = np.empty((N,N))

for i in range(N):
    R_f[i] = om_f[i]/om_dopp

plt.contourf(X,Y,R_f,50)
plt.title("Fast magnetosonic frequency")
plt.xlabel('Solar wind velocity $V_{SW}$')
plt.ylabel('angle k-B (in rad)')
plt.colorbar()
plt.show()
```

### Slow Magnetosonic mode

$\frac{\omega_{\text{S-MS}}}{\omega_{\text{Doppler}}}=\frac{\frac{k}{\sqrt{2}}\sqrt{\big(V_A^2 + C_s^2 - \sqrt{(V_A^2 + C_s^2)^2 - 4(V_A^2 C_s^2\frac{k_\parallel^2}{k^2})} \big)}}{kV_{\text{SW}}\cos(\theta_{k-SW})}=\frac{\frac{1}{\sqrt{2}}\sqrt{\big(V_A^2 + C_s^2 - \sqrt{(V_A^2 + C_s^2)^2 - 4(V_A^2 C_s^2\frac{k_\parallel^2}{k^2})} \big)}}{V_{\text{SW}}\cos(\theta_{k-SW})}$.

```python
N = 100
va  = 100
V_SW = np.linspace(400,800,N)
theta_kb = np.linspace(0,np.pi,N)
theta_kV = 40*2*np.pi/360
Beta = 2
gamma = 5/3

X,Y = np.meshgrid(V_SW,theta_kb)

cs = np.sqrt(Beta*gamma)*Va

om_s = np.sqrt((cs**2 + va**2)/2*(1 - np.sqrt(1 - 4* np.cos(theta_kb)**2*va**2*cs**2/(cs**2 + va**2)**2)))

om_dopp = V_SW*np.cos(theta_kV)

R_s = np.empty((N,N))

for i in range(N):
    R_s[i] = om_s[i]/om_dopp

plt.contourf(X,Y,R_s,50)
plt.title("slow magnetosonic frequency")
plt.xlabel('Solar wind velocity $V_{SW}$')
plt.ylabel('angle k-B (in rad)')
plt.colorbar()
plt.show()
```
