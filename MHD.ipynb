## Ideal MagnetoHydrodynamics (MHD)

In astrophysical plasma, it is well-known that there is multiple source of waves in the electricomagnetism spectrum, depending on the physical conditions considered.

Recall that under a static magnetic field $\mathbf{B_0}$, charged particles spirals around magnetic field lines. It is quite trivial to show that frequency is related to its amplitude $B$ via the formula: $\omega_i=\frac{ZeB_0}{m_ic} \ll \omega_e=\frac{eB_0}{m_e c}$ since $m_i \approx 2000m_e \gg m_e$. 

To show this, consider the linear compressible MHD model. Put yourself in the plasma frame, such that the overall plasma velocity is null. The plasma is supposed to be inviscid ($\nu=0$), an ideal conductor ($\eta=0$) and under a strong static magnetic field $B_0\hat z$. Also, assume the plasma undergo a reversible and adiabatic process. This assumption provides a state equation: Laplace's law. Laplace's law simply states that $\frac{p}{\rho^\gamma}=\text{Cte} \iff \frac{d}{dt}\frac{p}{(mn)^\gamma}=0$, where $\rho=mn= \rho_e + \rho_i$ and $\gamma=\frac{C_p}{C_v}$ is the adiabatic coefficient. Hence, the ideal MHD equations are:
$$
\begin{array} {cc}
\partial_t n + \nabla \cdot (nv) = 0 & \text{(Mass conservation equation)}\\
mn(\partial_t v + (v \cdot \nabla) v) = -\nabla p + J \times B & \text{(Momentum conservation equation)}\\
\partial_t B = -\nabla \times E & \text{(Maxwell-Faraday equation)}\\
\nabla \times B = \mu_0 J & \text{(Maxwell-Ampère equation in low frequency)}\\
\frac{d}{dt}(\frac{p}{(nm)^\gamma}) & \text{(Laplace's Law)}\\
E=-v \times B & \text{(Ohm's law)}.
\end{array}
$$
$J = ne(v_i-v_e)$ is the current density. After injecting Ohm's law into Maxwell-Faraday equation and Maxwell-Ampère into the momentum conservation equation, the set of equations for Ideal MHD is:
$$
\begin{array} {cc}
\partial_t n + \nabla \cdot (nv) = 0 & \text{(Mass conservation equation)}\\
mn(\partial_t v + (v \cdot \nabla) v) = -\nabla p + \frac1{\mu_0} (\nabla \times B) \times B & \text{(Momentum conservation equation)}\\
\partial_t B = \nabla \times (v \times B) & \text{(Maxwell-Faraday equation)}\\
\frac{d}{dt}(\frac{p}{(nm)^\gamma}) & \text{(Laplace's Law)}\\
\end{array}
$$

This form of Maxwell-Faraday equation unfolds from generalized Ohm's law in ideal conducting fluid and original Faraday equation: $\nabla \times E = -\partial_t B$. This set of equation are quite a good approximation for astrophysical plasma, where large-scale phenomena on the magnitude of $100 \text{km/h}$ have the most impact. For small scale physics though, it does not yield physically consistent results, especially around magnetic reconnection zones. The freezing theorem, a natural consequence of the ideal MHD equations from above, does not prevail in those conditions. From now on, it is assumed calculations are done in the plasma frame, that is, $v=0$.

Now, without loss of generality, take a linear perturbation of the magnetic field of the form: $B=B_0 \hat z + B_1= B_0 \hat z + \mathbf{B_1}e^{j(k_{\perp} z + k_{\parallel} x - \omega t)}$ with $|\mathbf{B_1}| \ll B_0$ and assume that $|\mathbf{v_1}| \ll 1, p_1 \ll p_0$ and $n_1 \ll n_0$.

After injection and equating orders, it yields:
$$
\begin{array} {cc}
\partial_t n_1 + n_0\nabla \cdot v_1 = 0 & \text{(Mass conservation equation)}\\
mn_0\partial_t v_1 = -\nabla p_1 + \frac1{\mu_0} (\nabla \times B_1) \times B_0 & \text{(Momentum conservation equation)}\\
\partial_t B_1 = \nabla \times (v_1 \times B_0) & \text{(Maxwell-Faraday equation)}\\
\partial_t p_1 = \frac{\gamma p_0}{mn_0} \partial_t n_1 & \text{(Laplace's Law)} \\
\end{array}
$$
Note that one has $4$ equations for $4$ unknowns $(p_1, n_1, v_1, B_1)$.
Note $C_s^2=\frac{\gamma p_0}{mn_0}=\sqrt{\beta \gamma} V_A$ the sound celerity, where $V_A=\frac{B_0}{\sqrt{\mu_0 m n_0}}$ and $\beta=\frac{2C_s^2}{\gamma V_A^2}$.
We would like to find relations between $\omega$, $k_\perp$ and $k_\parallel$ to have non-trivial solutions $\mathbf{B_1}$. In Fourier space, Laplace's law is: $\tilde p_1=\frac{\gamma p_0}{n_0}\tilde n_1$, mass conservation equation writes: $\tilde n_1= \frac{n_0 k \cdot \tilde v_1}{\omega}$ and Maxwell-Faraday's equation gives: $\tilde B_1= -\frac{k \times (\tilde v_1 \times B_0)}{\omega}$. Combining the two first ones establish $\tilde p_1=\frac{\gamma p_0}{\omega}k \cdot \tilde v_1 \implies \nabla p_1=jk \tilde p_1=jk \frac{\gamma p_0}{\omega} k \cdot \tilde v_1$. Momentum conservation equation gives: $-j\omega mn_0 \tilde v_1=-jk \frac{\gamma p_0}{\omega} k \cdot \tilde v_1 - \frac{j}{\mu_0 \omega} (k \times (k \times (\tilde v_1 \times B_0))) \times B_0$. After simplifications, it returns: 
$$
\boxed{\omega^2 \tilde v_1=k C_s^2 k \cdot \tilde v_1 + V_A^2 (k \times (k \times (\tilde v_1 \times \hat z))) \times \hat z}.
$$
Note that $\tilde v_1 \times \hat z=\begin{bmatrix} \tilde v_1^y \\ -\tilde v_1^x \\ 0 \end{bmatrix}$ and thus $k \times (\tilde v_1 \times \hat z)=\begin{bmatrix} k_\parallel \tilde v_1^x \\ k_\parallel \tilde v_1^y \\ -k_\perp \tilde v_1^x \end{bmatrix}$.
We now have to decompose this equation on each axis $\hat x$, $\hat y$ and $\hat z$:
- On $\hat z$. The second term of the r.h.s vanishes. It remains $\omega^2 \tilde v_1\cdot \hat z= C_s^2 k \cdot \tilde v_1 k \cdot \hat z=C_s^2 k \cdot \tilde v_1 k_\parallel \iff C_s^2 k \cdot \tilde v_1 k_\parallel - \omega^2 \tilde v_1 \cdot \hat z=0.$
- On $\hat y$. This time, the first term of the r.h.s vanishes because $k \cdot \hat y=0$. From the second term, only the $x$ components of $k \times (k \times (\tilde v_1 \times \hat z))$ remains. Hence, $\omega^2 \tilde v_1 \cdot \hat y= V_A^2 k_\parallel^2 \tilde v_1^y$.
- On $\hat x$. It immediately gives: $\omega^2 \tilde v_1 \cdot \hat x=k_\perp C_s^2 k \cdot \tilde v_1 - V_A^2 k^2 \tilde v_1^x$.

Thus, after eliminating $\tilde p_1$, $\tilde B_1$ and $\tilde n_1$, you arrive at the following linear equation:
$$
\underbrace{\begin{bmatrix}
-\omega^2 + (k V_A)^2 + k_\perp^2 C_s^2 & 0 & k_\perp k_\parallel C_s^2 \\
0 & -\omega^2 + k_\parallel^2 V_A^2 & 0 \\
k_\perp k_\parallel C_s^2 & 0 & -\omega^2 + k_\parallel^2 C_s^2
\end{bmatrix}}_{A \in \mathbb{R}^{3\times 3}} \tilde v_1 = 0
$$
for which $\text{Ker}(A) \neq \emptyset \iff \boxed{\text{det}(A) = 0}$. Each relation between $\omega$ and $k=(k_\perp, 0, k_\parallel)=k_\perp \hat x + k_\parallel \hat z$ such that $\text{det}(A)=0$ is called a **dispersion relation** and yields its own type of electromagnetic wave.

Just by looking at it, one can enumerates three solutions:
- If the middle term of the second column is null, then the whole column is null and obviously $\text{det}(A)=0$. It implies $\omega^2=k_\parallel^2V_A^2$. This is called Alfvén mode.
- If the first column and third column are colinears, then once again $\text{det}(A)=0$. It would implies that $\frac{-\omega^2 + (kV_A)^2 + k_\perp^2 C_s^2}{k_\perp k_\parallel C_s^2} = \frac{k_\perp k_\parallel C_s^2}{-\omega^2 + k_\parallel C_s^2} \iff (-\omega^2 + k_\parallel C_s^2)(-\omega^2 + (kV_A)^2 + k_\perp^2 C_s^2)=(k_\perp k_\parallel C_s^2)^2$. It admits two solutions: $\omega^2=\frac{k^2}2\big(V_A^2 + C_s^2 + \sqrt{(V_A^2 + C_s^2)^2 - 4(V_A^2 C_s^2\frac{k_\parallel^2}{k^2})} \big)$ and $\omega^2=\frac{k^2}2\big(V_A^2 + C_s^2 + \sqrt{(V_A^2 + C_s^2)^2 + 4(V_A^2 C_s^2\frac{k_\parallel^2}{k^2})} \big)$. Those are called respectively slow and fast magnetosonic modes.

To sum up, solutions are:
- **Alfvén mode**: Let $\omega^2= k_\parallel^2 V_A^2$. They are low-frequency i.e. lower than the ion gyrofrequency. This mode does not provoke density fluctuations at first order solution since $v_1 \in \text{Ker}(A)=\text{Vect}(\hat y)$ and continuity equation in frequency domain yields: $n_1=n_0\frac{k\cdot v_1}{\omega}=0$. It propagates at Alfvén velocity: $v_{\text{A}}^g=\frac{B_0}{\sqrt{\mu_0 \rho}}$. This is a group velocity (i.e. the actual moving speed of a wave packet), the phase velocity being: $v_{\text{A}}^p=\displaystyle\frac{v_{\text{A}}^g}{\sqrt{1 + \frac{v_{\text{A}}^g}{c^2}}}$. Since $v_A^g$ does not depend explicitly on $\omega$, the wave frequency, it is dispersionless.

- **fast compressible magnetosonic mode**: Let $\omega^2=\frac{k^2}2\big(V_A^2 + C_s^2 + \sqrt{(V_A^2 + C_s^2)^2 - 4(V_A^2 C_s^2\frac{k_\parallel^2}{k^2})} \big)$.
- **slow compressible magnetosonic mode**: Let $\omega^2=\frac{k^2}2\big(V_A^2 + C_s^2 - \sqrt{(V_A^2 + C_s^2)^2 - 4(V_A^2 C_s^2\frac{k_\parallel^2}{k^2})} \big)$.

Recall the definition of the phase speed: $v_\phi\equiv\frac{\omega}{\text{Re}(k)}$ and group speed $v_g\equiv\frac{\partial \omega}{\partial k}$. 

In the case of Alfvén waves, note that the Poynting vector $\Pi=\frac{E \times B}{\mu_0}=\frac{E_1 \times B_1}{\mu_0}$ is along $\hat z$, the energy propagates along the magnetic field lines.

## Objective

Consider a spacecraft with relative speed $V_{\text{SW}}$ to the Solar Wind.

In general, such a spacecraft probing an electromagnetic wave with angular frequency $\omega_{\text{plasma}}$ will have to take into account its own speed in the measurement, since it inevitably causes a Doppler shift $\omega_{\text{Doppler}}=k\cdot V_{\text{SW}}$ in the overall spectrum. At the end, the spectrum $\omega_{\text{sc}}$ measured by a probe is the sum of both: $\omega_{\text{sc}}=\omega_{\text{plasma}} + \omega_{\text{Doppler}} = \omega_{\text{plasma}} + k\cdot V_{\text{SW}}$. Note that $k$ is determined by $\omega_{\text{plasma}}$ via the dispersion relation, the latter being dependent on the wave considered. As the above section hints, we will consider the three wave discussed: Alfvén, slow and fast magnetosonic modes.

The objective of the next cells is to compare the influence of solar wind $V_{SW}$ with electromagnetic waves (Alfvén, Slow and Fast) propagating along $\textbf{k}$. Note $\theta_{k-B_0}$ the angle between the perturbation propagation and the static magnetic field $\mathbf{B_0}$. Note $\theta_{k-V_{SW}}$ the angle between $\textbf{k}$ and $\mathbf{V_{SW}}$. Therefore, $k_\parallel=k\cos(\theta_{k-B_0})$.

Taylor's hypothesis states that doppler shift completely dominates the spectrum observed by the spacecraft: $\omega_{\text{Doppler}}=kV_{\text{SW}}\cos(\theta_{kV_{\text{SW}}}) \gg \omega_{\text{plasma}}$.

Here, one would like to see under which configurations a doppler shift prevail over the other. A colormap will display the ratio $\frac{\omega_{\text{wave}}}{\omega_{\text{Doppler}}}$.

### Physical considerations for a spacecraft in Solar Wind

- $V_A=100 \, \text{km/s}$,

- $V_{SW}= 400-800 \, \text{km/s}$,

- $\theta_{k-SW}=40°$,

- $\theta_{k-B_0} = 0 - 180 °$

- $\beta=\frac{2C_s^2}{\gamma V_A^2}=2$

### Alfvén mode

$\frac{\omega_{\text{A}}}{\omega_{\text{Doppler}}}=\frac{kV_{\text{A}}\cos(\theta_{k-B_0})}{kV_{\text{SW}}\cos(\theta_{k-SW})}=\frac{V_{\text{A}}\cos(\theta_{k-B_0})}{V_{\text{SW}}\cos(\theta_{k-SW})}$

### Fast Magnetosonic mode

$\frac{\omega_{\text{F-MS}}}{\omega_{\text{Doppler}}}=\frac{\frac{k}{\sqrt{2}}\sqrt{\big(V_A^2 + C_s^2 + \sqrt{(V_A^2 + C_s^2)^2 - 4(V_A^2 C_s^2\frac{k_\parallel^2}{k^2})} \big)}}{kV_{\text{SW}}\cos(\theta_{k-SW})}=\frac{\frac{1}{\sqrt{2}}\sqrt{\big(V_A^2 + C_s^2 + \sqrt{(V_A^2 + C_s^2)^2 - 4(V_A^2 C_s^2\cos^2(\theta_{k-B_0}))} \big)}}{V_{\text{SW}}\cos(\theta_{k-SW})}$.

### Slow Magnetosonic mode

$\frac{\omega_{\text{S-MS}}}{\omega_{\text{Doppler}}}=\frac{\frac{k}{\sqrt{2}}\sqrt{\big(V_A^2 + C_s^2 - \sqrt{(V_A^2 + C_s^2)^2 - 4(V_A^2 C_s^2\frac{k_\parallel^2}{k^2})} \big)}}{kV_{\text{SW}}\cos(\theta_{k-SW})}=\frac{\frac{1}{\sqrt{2}}\sqrt{\big(V_A^2 + C_s^2 - \sqrt{(V_A^2 + C_s^2)^2 - 4(V_A^2 C_s^2\cos(\theta_{k-B_0})^2)} \big)}}{V_{\text{SW}}\cos(\theta_{k-SW})}$.

```python
#Number of entries
N = 100

#Alfvén speed (km/s)
Va = 100

#Solar wind array of speed
V_SW = np.linspace(400,800,N)

#Angle between magnetic field and wave
theta_kb = np.linspace(0,np.pi,N)
theta_kV = 40*2*np.pi/360
Beta = 2
gamma = 5/3
Cs = np.sqrt(Beta*gamma)*Va/2

om_a = Va*np.abs(np.cos(theta_kb))
om_f = np.sqrt((Cs**2 + Va**2)/2*(1 + np.sqrt(1 - (2*np.cos(theta_kb)*Va*Cs)**2/(Cs**2 + Va**2)**2)))
om_s = np.sqrt((Cs**2 + Va**2)/2*(1 - np.sqrt(1 - (2*np.cos(theta_kb)*Va*Cs)**2/(Cs**2 + Va**2)**2)))

X,Y = np.meshgrid(V_SW,theta_kb)

om_dopp = V_SW*np.cos(theta_kV)

R_f=np.empty((N,N))
R_s=np.empty((N,N))
R=np.empty((N,N))

for i in range(N):
        R[i]=om_a[i]/om_dopp
        R_f[i]=om_f[i]/om_dopp
        R_s[i]=om_s[i]/om_dopp

fig, ax=plt.subplots(nrows=3, ncols=1, figsize=(7,20))

mappa=ax[0].contourf(X,Y,R,50)
ax[0].set_title('Alfvén wave')
ax[0].set_xlabel('Solar wind velocity $V_{SW}$')
ax[0].set_ylabel('angle k-B (in rad)')
plt.colorbar(mappa)

mappa=ax[1].contourf(X,Y,R_f,50)
ax[1].set_title("Fast magnetosonic frequency")
ax[1].set_xlabel('Solar wind velocity $V_{SW}$')
ax[1].set_ylabel('angle k-B (in rad)')
plt.colorbar(mappa)

mappa=ax[2].contourf(X,Y,R_s,50)
ax[2].set_title("slow magnetosonic frequency")
ax[2].set_xlabel('Solar wind velocity $V_{SW}$')
ax[2].set_ylabel('angle k-B (in rad)')
plt.colorbar(mappa)

plt.show()
```

## Incompressible Hall-MHD

We consider this time another model for MHD. The last model can not explain a variety of phenomena, such as magnetic line breaking and magnetic reconnection. To take into account this physics, a new term must be added in Ohm's law, the Hall's term. It can be shown by a dimensional analysis that it describes spatial scales comparables to the ion inertial length, that is, we adopt a more fine-grained description of the MHD fluid. The equations are:
$$
\begin{array} {cc}
\partial_t n + \nabla \cdot (nv) = 0 & \text{(Mass conservation equation)}\\
\nabla \cdot v = 0 & \text{(Incompressible fluid)}\\
mn(\partial_t v + (v \cdot \nabla) v) = -\nabla p + J \times B & \text{(Momentum conservation equation)}\\
\partial_t B = \nabla \times E & \text{(Maxwell-Faraday equation)}\\
\nabla \times B = \mu_0 J & \text{(Maxwell-Ampère equation in low frequency)} \\
\frac{d}{dt}(\frac{p}{(nm)^\gamma}) & \text{(Laplace's Law)}\\
E=-v \times B + \underbrace{\frac1{en}J \times B}_{\text{Hall's term}} & \text{(Ohm's law)}.
\end{array}
$$

Combining Ohm's law with Maxwell-Faraday equation yields: $\partial_t B=\nabla \times (v \times B) - \frac1{en}\nabla \times (J\times B)$. Only the second term of the r.h.s is new here. By taking the Alfvén units: $b=\frac{B}{\sqrt{\mu_0 \rho_0}}$ and $j=\sqrt{\frac{\mu_0}{\rho_0}}J=\nabla \times b$, the dynamics of the magnetic equation is: $\partial_t b = \nabla \times (v \times b) - d_i \nabla \times (j \times b)$. $d_i$ is the ion inertial length, equals to: $d_i\equiv\frac{c}{\omega_{pi}}$. A quick dimension analysis shows that if $d_i \ll L$ ($L$ is a characteristic scale of the system), then the Hall term can be neglected.

Take once again a linear perturbation of the magnetic field of the form: $B=B_0 \hat z + B_1= B_0 \hat z + \mathbf{B_1}e^{j(k_{\perp} z + k_{\parallel} x - \omega t)}$ with $|\mathbf{B_1}| \ll B_0$ and assume that $|\mathbf{v_1}| \ll 1, p_1 \ll p_0$ and $n_1 \ll n_0$.

After injection and equating orders, it yields:
$$
\begin{array} {cc}
\partial_t n_1 + n_0\nabla \cdot v_1 = 0 & \text{(Mass conservation equation)}\\
\nabla \cdot v_1 = 0 & \text{(Incompressible fluid)}\\
mn_0\partial_t v_1 = -\nabla p_1 + \frac1{\mu_0} (\nabla \times B_1) \times B_0 & \text{(Momentum conservation equation)}\\
\partial_t B_1 = \nabla \times (v_1 \times B_0) - \frac1{en_0 \mu_0}\nabla \times ((\nabla \times B_1) \times B_0)& \text{(Maxwell-Faraday equation)}\\
\partial_t p_1 = \frac{\gamma p_0}{mn_0} \partial_t n_1 & \text{(Laplace's Law)} \\
\end{array}
$$
As before, we have $4$ equations for $4$ unknowns. By taking linear perturbations of the form $\propto e^{j(k_{\parallel} z + k_\perp x - \omega t)}$, all the equations are the same as above, with the exception of Ohm's law, that now reads in Fourier space:
$-\omega \tilde B_1=k \times (\tilde v_1 \times B_0) - \frac1{en_0\mu_0} j k \times ((k \times \tilde B_1) \times B_0)$ and incompressibility: $k \cdot \tilde v_1=0$. I eliminate this time $\tilde v_1$, $\tilde p_1$ and $\tilde n_1$ to have only $\tilde B_1$ left in the Maxwell-Faraday equation. Momentum conservation equation reads:
$$
\omega^2 mn_0 \tilde v_1= \underbrace{k C_s^2 \tilde v_1 \cdot k}_{=0} + \frac{\omega}{\mu_0} (k \times \tilde B_1) \times B_0 \implies \boxed{\tilde v_1=\frac{1}{mn_0\mu_0\omega} (k \times \tilde B_1) \times B_0}
$$

One can use the identity for all vector fields $A$ and $B$: $\nabla \times (A \times B)= (B \cdot \nabla) A - (A \cdot \nabla)B + A(\nabla \cdot B) - B(\nabla \cdot A)$. For $A=v_1$ and $B=B_0$ a uniform vector field, it reduces to: $\nabla \times (v_1 \times B_0)=(B_0 \cdot \nabla) v_1 - B_0(\nabla \cdot v_1)$. In Fourier space, $\boxed{k \times (\tilde v_1 \times B_0)=B_0 k_\parallel \tilde v_1 - \underbrace{B_0(k \cdot \tilde v_1)}_{=0}=\frac{B_0 k_\parallel}{\mu_0 m n_0 \omega} (k \times \tilde B_1) \times B_0}$. The same formula for $A=\nabla \times B_1$ and $B=B_0$ yields: $\nabla \times ((\nabla \times B_1) \times B_0)= (B_0 \cdot \nabla) (\nabla \times B_1) - B_0 \underbrace{\nabla \cdot (\nabla \times B_1)}_{=0}$ because the divergence of a rotational is always null. In Fourier space, $jk \times ((k \times B_1) \times B_0)= jB_0 k_\parallel ( k \times B_1)$.

Plugging this into Maxwell-Faraday equation reads:
$$
\begin{array}{ccc}
-\omega^2 \tilde B_1 &= \frac{B_0^2 k_\parallel}{\mu_0 m n_0} (k \times \tilde B_1) \times \hat z - \frac{\omega B_0j k_\parallel}{e n_0 \mu_0} (k \times \tilde B_1) \\
&= V_A^2 k_\parallel (k \times \tilde B_1) \times \hat z - \frac{V_A j k_\parallel \omega \sqrt{m}}{e\sqrt{n_0 \mu_0}}(k \times \tilde B_1) & V_A=\frac{B}{\sqrt{mn_0 \mu_0}}\\
&= V_A^2 k_\parallel (k \times \tilde B_1) \times \hat z - \frac{V_A j k_\parallel c \omega \sqrt{m \epsilon_0}}{e\sqrt{n_0}}(k \times \tilde B_1) & c=\frac{1}{\sqrt{\mu_0 \epsilon_0}}\\
&= V_A^2 k_\parallel (k \times \tilde B_1) \times \hat z - \frac{V_A j k_\parallel c \omega}{\omega_{pi}}(k \times \tilde B_1) & \omega_{pi}=\sqrt{\frac{n_0e^2}{m \epsilon_0}} \\
&= V_A^2 k_\parallel (k \times \tilde B_1) \times \hat z - V_A j k_\parallel d_i \omega(k \times \tilde B_1) & d_i=\frac{c}{w_{pi}}\\
\end{array}
$$

I shall now proceed as above, and project on each axis. Note that $k \times \tilde B_1 = \begin{bmatrix} -k_\parallel \tilde{B}_1^y \\ k_\parallel \tilde{B}_1^x - k_\perp \tilde{B}_1^z \\ k_\perp \tilde{B}_1^y \end{bmatrix}$ and $(k \times \tilde B_1) \times \hat z=\begin{bmatrix} k_\parallel \tilde{B}_1^x - k_\perp \tilde{B}_1^z \\ -k_\perp \tilde{B}_1^y \\ 0 \end{bmatrix}$.

- On $\hat z$. The first of the r.h.s is null. It reduces to: $\omega^2 \tilde B_1^z=V_A j k_\parallel d_i \omega(k \times \tilde B_1) \cdot \hat z= V_A j k_\parallel d_i \omega k_\perp \tilde{B_1}^y$.
- On $\hat y$. It follows: $\omega^2 \tilde B_1^y= V_A^2 k_\perp k_\parallel \tilde B_1^y + V_A j k_\parallel d_i \omega (k_\parallel \tilde{B}_1^x - k_\perp \tilde{B}_1^z)$.
- On $\hat x$. It follows $\omega^2 \tilde B_1^x=V_A^2 k_\parallel (k_\perp \tilde{B}_1^z - k_\parallel \tilde{B}_1^x) - j V_A \omega d_i k_\parallel^2 \tilde B_1^y$.

Written with matrix algebra, it amounts to:
$$
\underbrace{\begin{bmatrix}
-\omega^2 - (k_\parallel V_A)^2 & -j V_A \omega d_i k_\parallel^2 & k_\perp k_\parallel V_A^2 \\
jV_A d_i \omega k_\parallel^2 & -\omega^2 + k_\parallel k_\perp V_A^2 & - jV_A d_i \omega k_\parallel k_\perp \\
0 & V_A j k_\parallel d_i \omega k_\perp & -\omega^2
\end{bmatrix}}_{K \in \mathbb{C}^{3\times 3}} \tilde B_1 = 0
$$

Let's compute $\text{det}(K)$ and solve $\text{det}(K)=0$. By expanding the last line, one has $\text{det}(K)=\omega^2((\omega^2 + (k_\parallel V_A)^2)(-\omega^2 + k_\parallel k_\perp V_A^2) + V_A^2 \omega^2 d_i^2 k_\parallel^4) - jV_A \omega d_i k_\parallel k_\perp(jV_A \omega d_i k_\parallel^2(\omega^2 + (k_\parallel V_A)^2) - jk_\perp k_\parallel^3 V_A^3 d_i \omega )$.

Assume, despite the loss of generality, that $k_\perp=0 \iff k=k_\parallel$. Second term vanishes and thus, $\text{det}(K)=\omega^2(-(\omega^2 + (k V_A)^2)\omega^2 + V_A^2 \omega^2 d_i^2 k^4) = 0 \iff -(\omega^2 + (k V_A)^2)\omega^2 + V_A^2 \omega^2 d_i^2 k^4=0 \iff \boxed{(k V_A)^2(d_i^2 k^2 - 1) =\omega^2}$. A existence condition appears, $kd_i \geq 1 \iff d_i \geq \frac1k$, that is, the scale of interest must be smaller than ion inertial length, which is coherent with our previous discussion on the Hall term.
